# Stop Word Removal

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = stopwords.words('english')
stop_words[:15]
```

```python
from nltk.tokenize import word_tokenize
text = "This is an example to demonstrate removing stop words." 
tokens = word_tokenize(text)
tokens
```

```python
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
filtered_tokens
```

# punctuation Removal

```python
import string 
punctuations = string.punctuation
punctuations
```

```python
clean_tokens = [word for word in filtered_tokens if word not in string.punctuation]
clean_tokens
```

### Stemming 
Reduce words to their root form by chopping off suffixes or prefix 
* running --> run
* reading --> read

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in clean_tokens]
stemmed_tokens
```

### Lemmatization
Reduce words to their base form
organizations --> organization

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in clean_tokens]
lemmatized_tokens
```
